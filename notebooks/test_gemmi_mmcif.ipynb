# Test GEMMI

```python
import gemmi
```

### 1A0G

```python
st = gemmi.read_structure("/nfs/scratch/pdb_dimers/assemblies/1a0g-assembly1.cif.gz")

for model in st:
    print("model found")
    for chain in model:
        print("Chain: ", chain.name)
        for res in chain:
            print("Residue: ", res.name, res.seqid)
```

```python
st.setup_entities()
for entity in st.entities:
    print(entity.name)
    print(entity.entity_type)
    if entity.name == "1":
        seq_3letter = entity.full_sequence
        seq_1letter = gemmi.one_letter_code(seq_3letter)
        print("3-letter code: ", seq_3letter)
        print("1-letter code: ", seq_1letter)
```
