```python
import pandas as pd
import os
import matplotlib as plt
```

```python
base_dir = "/nfs/scratch/pdb_dimers"
selected_assemblies = pd.read_csv(os.path.join(base_dir, "selected_assemblies.tsv"), sep="\t")

selected_assemblies.head()
```

### Homo vs Hetero

```python
dimer_counts = selected_assemblies["dimer_type"].value_counts()

ax = dimer_counts.plot.pie(
    autopct="%1.1f%%",
    figsize=(5, 5),
    ylabel="",
    title="Homo vs Hetero Dimers"
)
ax.set_aspect("equal")
```
